```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import balanced_accuracy_score, roc_auc_score, make_scorer
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
import xgboost as xgb
import matplotlib as plt
```

```python
df = pd.read_csv('../data/creditcard.csv')
```

```python
df = pd.get_dummies(df, columns=["type"], dtype=int)
```

```python
df = pd.get_dummies(df, columns=["type"], dtype=int)
df["dest_is_merchant"] = df["nameDest"].str.startswith("M").astype(int)
df = df.drop(columns=["nameOrig", "nameDest"])
```

```python
cutoff = df["step"].quantile(0.8)          
train = df[df["step"] <= cutoff]
test  = df[df["step"] >  cutoff]
train.head()
```

```python
X_train = train.drop(columns = ["isFraud"]).copy()
X_test = test.drop(columns = ["isFraud"]).copy()
y_train = train['isFraud'].copy()
y_test = test['isFraud'].copy()
```

```python
ratio = (y_train == 0).sum() / (y_train == 1).sum()

clf_xgb = xgb.XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    early_stopping_rounds=50,
    scale_pos_weight=ratio,
    learning_rate=0.03,
    n_estimators=1000,
    max_depth=6,
    min_child_weight=5,
    subsample=0.8,
    colsample_bytree=0.8,
    seed=42,
)
```

```python
clf_xgb.fit(X_train, y_train, verbose=True, eval_set=[(X_test, y_test)])
```

```python
print(clf_xgb.best_iteration)                        
print(clf_xgb.get_booster().num_boosted_rounds()) 
```

```python
ConfusionMatrixDisplay.from_estimator(clf_xgb, X_test, y_test, display_labels=['Legit','Fraud'])
```

```python
clf_xgb.save_model("../artifacts/heimdall_fraud_model.json")
```
